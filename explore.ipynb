```python
import pandas as pd 
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df_train_x = pd.read_csv('train.csv')
df_train_x = df_train_x.drop(['PassengerId', 'Name', 'Age','Ticket', 'Cabin'], axis=1)
# scaler = preprocessing.MinMaxScaler()
# df_train_x[['Fare']] = scaler.fit_transform(df_train_x[['Fare']])
df_train_y = df_train_x.filter(['Survived'], axis=1)
df_train_x = df_train_x.drop('Survived', axis=1)
head = {}
count = 0
for value in df_train_x['Embarked']:
    if value not in head:
        head[value] = count
        count+=1
print(head)
df_train_x['Embarked'] = df_train_x['Embarked'].map(head)
df_train_x['Sex'] = df_train_x['Sex'].apply(lambda x: 1 if x == 'male' else 0)
print(df_train_x)
```

```python
clf = RandomForestClassifier(n_estimators = 100)
```

```python
train_x = df_train_x.values
train_y = df_train_y.values.ravel()
clf.fit(train_x, train_y)
training_accuracy = clf.score(train_x, train_y)
print(f"Training Accuracy: {training_accuracy * 100:.2f}%")
```

```python
df_test_x = pd.read_csv('test.csv')
ids = df_test_x.filter(['PassengerId'], axis=1)
df_test_x = df_test_x.drop(['PassengerId','Name', 'Age','Ticket', 'Cabin'], axis=1)
# scaler = preprocessing.MinMaxScaler()
# df_train_x[['Fare']] = scaler.fit_transform(df_train_x[['Fare']])
head = {}
count = 0
for value in df_test_x['Embarked']:
    if value not in head:
        head[value] = count
        count+=1
print(head)
df_test_x['Embarked'] = df_test_x['Embarked'].map(head)
df_test_x['Sex'] = df_test_x['Sex'].apply(lambda x: 1 if x == 'male' else 0)
test_y = clf.predict(df_test_x.values)
output = pd.DataFrame({'PassengerId': ids['PassengerId'], 'Survived': test_y})
output.to_csv('submission.csv', index=False)
```

```python
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(train_x, train_y)
```

```python
from sklearn.svm import SVC

# Initialize and train SVM (with an RBF kernel as an example)
svm_clf = SVC(kernel='rbf', random_state=42)
svm_clf.fit(train_x, train_y)

# Evaluate on test data
y_pred_svm = svm_clf.predict(train_x)
svm_accuracy = accuracy_score(train_y, y_pred_svm)
print(f"SVM Test Accuracy: {svm_accuracy * 100:.2f}%")
```
